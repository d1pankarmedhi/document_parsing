# Table parsing with PyMuPDF

```python
import fitz 
import os 
```

```python
def extract_tables(file_path:str):
    """
    Extract tables from pdfs.
    """
    doc = fitz.open(file_path)
    result = []

    for page in doc:
        tables = []
        headers = []

        for table in page.find_tables():
            tables.append(table.to_pandas())
            header_names = [name if name[:3] != "Col" else "" for name in table.header.names]
            headers.append(header_names)
        if len(tables) > 0:
            result.append({"tables": tables,"headers": headers, "page": page.number})
    return result

tables = extract_tables(os.path.join(os.getcwd(), "documents", "document_3.pdf"))
```

```python
# number of tables per page
print("\nNumber of tables per page:")
for table_info in tables:
    print(f"Page {table_info['page']} - {len(table_info['tables'])} table(s)")
```
